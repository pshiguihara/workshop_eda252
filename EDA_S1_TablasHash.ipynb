```python
def es_primo(n: int) -> bool:
    """
    Determina si un número entero es primo.

    Parámetros:
        n (int): número a verificar

    Retorna:
        bool: True si n es primo, False en caso contrario
    """
    if n <= 1:
        return False
    if n <= 3:
        return True
    if n % 2 == 0 or n % 3 == 0:
        return False

    # Verificación hasta la raíz cuadrada de n
    i = 5
    while i * i <= n:
        if n % i == 0 or n % (i + 2) == 0:
            return False
        i += 6
    return True
```

# Tabla Hash con Open Addressing

```python
class TablaHash(object):
  def __init__(self, tam=7, hash=hashSimple, probe=linearProbe, factorCargaMax=0.5):
    self.__tabla = [None]*tam
    self.__nItems = 0
    self.__hash = hash
    self.__probe = probe
    self.__factorCargaMax = factorCargaMax

  def __len__(self):
    return self.__nItems

  def celdas(self):
    return len(self.__tabla)

  def hash(self, clave):
    return self.__hash(clave) % self.celdas()

  def buscar(self, clave):
    i = self.__encontrar(clave)
    return (None if(i is None) or self.__tabla[i] is None or self.__tabla[i][0]!=clave else self.__tabla[i][1])

  __Eliminado = (None, 'Marcardor de eliminacion')

  def __encontrar(self, clave, eliminadoOk=False):
    for i in self.__probe(self.hash(clave), clave, self.celdas()):
      if(self.__tabla[i] is None or (self.__tabla[i] is TablaHash.__Eliminado and eliminadoOk) or self.__tabla[i][0]==clave):
        return i
    return None

  def insertar(self, clave, valor):
    i = self.__encontrar(clave, eliminadoOk=True)
    if i is None:
      raise Exception("Tabla Hash: secuencia falló en la inserción")
    if (self.__tabla[i] is None or self.__tabla[i] is TablaHash.__Eliminado):
      self.__tabla[i] = (clave, valor)
      self.__nItems += 1
      if self.factorCarga() > self.__factorCargaMax:
        self.__crecimientoTabla()
      return True

    if self.__tabla[i][0] == clave:
      self.__tabla[i] = (clave, valor)
      return False

  def factorCarga(self):
    return self.__nItems / len(self.__tabla)

  def __crecimientoTabla(self):
    tablaAntigua = self.__tabla
    tam = len(tablaAntigua)*2+1
    while not es_primo(tam):
      tam += 2
    self.__tabla = [None]*tam
    self.__nItems = 0
    for i in range(len(tablaAntigua)):
      if(tablaAntigua[i] and tablaAntigua[i] is not TablaHash.__Eliminado):
        self.insertar(*tablaAntigua[i])

  def eliminar(self, clave, ignoreMissing=False):
    i = self.__encontrar(clave)
    if (i is None or self.__tabla[i] is None or self.__tabla[i][0] != clave):
      if ignoreMissing:
        return
      raise Exception("No se puede eliminar la clave {} en la Tabla Hash".format(clave))
    self.__tabla[i] = TablaHash.__Eliminado
    self.__nItems -= 1

  def recorrer(self):
    for i in range(len(self.__tabla)):
      if(self.__tabla[i] and self.__tabla[i] is not TablaHash.__Eliminado):
        yield self.__tabla[i]
```

```python
# @title Prueba de Tabla Hash con Open Addressing
# ⬅️ CELDA 1: Panel interactivo para TablaHash (insertar y reportar)
from typing import Any, Generator
import ipywidgets as widgets
from IPython.display import display, Markdown, clear_output
import pandas as pd

# --- Hash y Probing ---
def hashSimple(clave: Any) -> int:
    if isinstance(clave, int):
        return clave & 0x7FFFFFFF
    if isinstance(clave, str):
        h = 0
        for ch in clave:
            h = (h * 257 + ord(ch)) & 0x7FFFFFFF
        return h
    if isinstance(clave, (tuple, list)):
        h = 0
        for x in clave:
            h = (h * 257 + hashSimple(x)) & 0x7FFFFFFF
        return h
    return hashSimple(repr(clave))

def linearProbe(i0: int, clave: Any, m: int):
    for k in range(m):
        yield (i0 + k) % m

def doubleHashProbe(i0: int, clave: Any, m: int):
    h2 = (hashSimple((clave, "h2")) % (m - 1)) + 1
    if h2 % 2 == 0:
        h2 += 1
    for k in range(m):
        yield (i0 + k * h2) % m

def es_primo(n: int) -> bool:
    if n <= 1: return False
    if n <= 3: return True
    if n % 2 == 0 or n % 3 == 0: return False
    i = 5
    while i * i <= n:
        if n % i == 0 or n % (i + 2) == 0:
            return False
        i += 6
    return True

# --- Tu clase TablaHash (sin cambios) ---
class TablaHash(object):
  def __init__(self, tam=7, hash=hashSimple, probe=linearProbe, factorCargaMax=0.5):
    self.__tabla = [None]*tam
    self.__nItems = 0
    self.__hash = hash
    self.__probe = probe
    self.__factorCargaMax = factorCargaMax

  def __len__(self):
    return self.__nItems

  def celdas(self):
    return len(self.__tabla)

  def hash(self, clave):
    return self.__hash(clave) % self.celdas()

  def buscar(self, clave):
    i = self.__encontrar(clave)
    return (None if(i is None) or self.__tabla[i] is None or self.__tabla[i][0]!=clave else self.__tabla[i][1])

  __Eliminado = (None, 'Marcardor de eliminacion')

  def __encontrar(self, clave, eliminadoOk=False):
    for i in self.__probe(self.hash(clave), clave, self.celdas()):
      if(self.__tabla[i] is None or (self.__tabla[i] is TablaHash.__Eliminado and eliminadoOk) or self.__tabla[i][0]==clave):
        return i
    return None

  def insertar(self, clave, valor):
    i = self.__encontrar(clave, eliminadoOk=True)
    if i is None:
      raise Exception("Tabla Hash: secuencia falló en la inserción")
    if (self.__tabla[i] is None or self.__tabla[i] is TablaHash.__Eliminado):
      self.__tabla[i] = (clave, valor)
      self.__nItems += 1
      if self.factorCarga() > self.__factorCargaMax:
        self.__crecimientoTabla()
      return True

    if self.__tabla[i][0] == clave:
      self.__tabla[i] = (clave, valor)
      return False

  def factorCarga(self):
    return self.__nItems / len(self.__tabla)

  def __crecimientoTabla(self):
    tablaAntigua = self.__tabla
    tam = len(tablaAntigua)*2+1
    while not es_primo(tam):
      tam += 2
    self.__tabla = [None]*tam
    self.__nItems = 0
    for i in range(len(tablaAntigua)):
      if(tablaAntigua[i] and tablaAntigua[i] is not TablaHash.__Eliminado):
        self.insertar(*tablaAntigua[i])

  def eliminar(self, clave, ignoreMissing=False):
    i = self.__encontrar(clave)
    if (i is None or self.__tabla[i] is None or self.__tabla[i][0] != clave):
      if ignoreMissing:
        return
      raise Exception("No se puede eliminar la clave {} en la Tabla Hash".format(clave))
    self.__tabla[i] = TablaHash.__Eliminado
    self.__nItems -= 1

  def recorrer(self):
    for i in range(len(self.__tabla)):
      if(self.__tabla[i] and self.__tabla[i] is not TablaHash.__Eliminado):
        yield self.__tabla[i]

# --- UI ---
tam_init   = widgets.IntText(value=7, description='Tamaño:', layout=widgets.Layout(width='200px'))
factor_max = widgets.FloatText(value=0.5, description='α max:', layout=widgets.Layout(width='180px'))
probe_sel  = widgets.Dropdown(options=[('Sondeo lineal','linear'),('Doble hashing','double')],
                              value='linear', description='Probe:', layout=widgets.Layout(width='240px'))
btn_reset  = widgets.Button(description='Crear/Reset', button_style='info')

key_in   = widgets.Text(value='', description='Clave:',  placeholder='ej. a, 42, (1,2)')
val_in   = widgets.Text(value='', description='Valor:',  placeholder='ej. 10, "hola"')
btn_ins  = widgets.Button(description='Insertar', button_style='success')

# (Opcional) Buscar/Eliminar, por si quieres probar:
key_find = widgets.Text(value='', description='Clave:')
btn_bus  = widgets.Button(description='Buscar')
btn_del  = widgets.Button(description='Eliminar', button_style='warning')

stats_out = widgets.Output()
table_out = widgets.Output()
log_out   = widgets.Output()

current = {'th': None}

def make_table():
    tam = max(3, int(tam_init.value))
    alpha = float(factor_max.value)
    probe = linearProbe if probe_sel.value == 'linear' else doubleHashProbe
    return TablaHash(tam=tam, hash=hashSimple, probe=probe, factorCargaMax=alpha)

def parse_literal(s: str):
    s = s.strip()
    if s == '': return ''
    import ast
    # Int directo
    if s.lstrip('+-').isdigit():
        try: return int(s)
        except: pass
    # Float
    try:
        f = float(s)
        return f
    except:
        pass
    # Literal seguro (tuplas/listas/dicts)
    try:
        return ast.literal_eval(s)
    except:
        return s  # como string

def refresh(th: TablaHash, msg=''):
    with stats_out:
        clear_output(wait=True)
        display(Markdown(f"**Items:** {len(th)} | **Celdas:** {th.celdas()} | **Factor de carga (α):** {th.factorCarga():.3f}"))
        if msg: display(Markdown(f"_Última acción:_ {msg}"))
    # Vista interna por índice
    tabla = getattr(th, "_TablaHash__tabla")
    rows = []
    TUMBA = getattr(TablaHash, "_TablaHash__Eliminado")
    for i, cell in enumerate(tabla):
        if cell is None:
            estado, clave, valor = "Vacío", None, None
        elif cell is TUMBA:
            estado, clave, valor = "Eliminado (tumba)", None, None
        else:
            estado, (clave, valor) = "Ocupado", cell
        rows.append({"Índice": i, "Estado": estado, "Clave": clave, "Valor": valor})
    df = pd.DataFrame(rows)
    with table_out:
        clear_output(wait=True)
        display(df.style.hide(axis='index'))

def on_reset(_):
    th = make_table()
    current['th'] = th
    with log_out:
        clear_output(wait=True)
        print(f"Tabla creada: tamaño={th.celdas()} | α max={factor_max.value} | Probe={probe_sel.label}")
    refresh(th, "Tabla creada/reset")

def on_insert(_):
    th = current.get('th') or make_table()
    current['th'] = th
    clave = parse_literal(key_in.value)
    valor = parse_literal(val_in.value)
    try:
        nuevo = th.insertar(clave, valor)
        msg = f"Insertar clave={repr(clave)}, valor={repr(valor)} " + ("(nuevo)" if nuevo else "(actualización)")
        with log_out:
            clear_output(wait=True)
            print(msg)
        refresh(th, msg)
    except Exception as e:
        with log_out:
            clear_output(wait=True)
            print("ERROR:", e)

def on_buscar(_):
    th = current.get('th')
    if th is None:
        with log_out:
            clear_output(wait=True)
            print("Primero crea la tabla.")
        return
    clave = parse_literal(key_find.value)
    val = th.buscar(clave)
    with log_out:
        clear_output(wait=True)
        print(f"Buscar clave={repr(clave)} ->", val)

def on_eliminar(_):
    th = current.get('th')
    if th is None:
        with log_out:
            clear_output(wait=True)
            print("Primero crea la tabla.")
        return
    clave = parse_literal(key_find.value)
    try:
        th.eliminar(clave)
        with log_out:
            clear_output(wait=True)
            print(f"Eliminar clave={repr(clave)} -> OK")
        refresh(th, f"Eliminar clave={repr(clave)}")
    except Exception as e:
        with log_out:
            clear_output(wait=True)
            print("ERROR:", e)

btn_reset.on_click(on_reset)
btn_ins.on_click(on_insert)
btn_bus.on_click(on_buscar)
btn_del.on_click(on_eliminar)

display(Markdown("### Configuración"))
display(widgets.HBox([tam_init, factor_max, probe_sel, btn_reset]))

display(Markdown("### Insertar"))
display(widgets.HBox([key_in, val_in, btn_ins]))

display(Markdown("### (Opcional) Buscar/Eliminar"))
display(widgets.HBox([key_find, btn_bus, btn_del]))

display(Markdown("### Resumen"))
display(stats_out)

display(Markdown("### Vista interna (índice, estado, clave, valor)"))
display(table_out)

display(Markdown("### Log"))
display(log_out)

# inicializa
on_reset(None)
```

# Tablas Hash con Separate Chaining

```python
from typing import Any, List, Tuple

def hashSimple(clave: Any) -> int:
    """
    Hash determinista para int, str, tuplas/listas; para otros objetos usa repr().
    Evita el hash aleatorio por sesión de Python.
    """
    if isinstance(clave, int):
        return clave & 0x7FFFFFFF
    if isinstance(clave, str):
        h = 0
        for ch in clave:
            h = (h * 257 + ord(ch)) & 0x7FFFFFFF
        return h
    if isinstance(clave, (tuple, list)):
        h = 0
        for x in clave:
            h = (h * 257 + hashSimple(x)) & 0x7FFFFFFF
        return h
    return hashSimple(repr(clave))

def es_primo(n: int) -> bool:
    if n <= 1: return False
    if n <= 3: return True
    if n % 2 == 0 or n % 3 == 0: return False
    i = 5
    while i * i <= n:
        if n % i == 0 or n % (i + 2) == 0:
            return False
        i += 6
    return True

# --- Implementación con Separate Chaining ---

class TablaHashChaining(object):
    """
    Tabla Hash con encadenamiento separado (buckets como listas).
    Interfaz alineada con tu TablaHash de direccionamiento abierto:
      - __len__(), celdas(), hash(clave), buscar(clave)
      - insertar(clave, valor) -> True si nuevo, False si actualización
      - eliminar(clave, ignoreMissing=False)
      - recorrer() -> genera (clave, valor) de todas las celdas ocupadas
      - factorCarga() = nItems / numBuckets
    """

    def __init__(self, tam: int = 7, hash=hashSimple, factorCargaMax: float = 0.75):
        if tam < 3: tam = 3
        # Ajustar a número primo para mejorar la dispersión
        if not es_primo(tam):
            t = tam if tam % 2 == 1 else tam + 1
            while not es_primo(t):
                t += 2
            tam = t
        self.__buckets: List[List[Tuple[Any, Any]]] = [[] for _ in range(tam)]
        self.__nItems = 0
        self.__hash = hash
        self.__factorCargaMax = factorCargaMax

    # --- Consultas básicas ---
    def __len__(self):
        return self.__nItems

    def celdas(self):
        """Número de buckets (no de elementos)."""
        return len(self.__buckets)

    def hash(self, clave):
        """Índice del bucket donde debería caer la clave."""
        return self.__hash(clave) % self.celdas()

    def factorCarga(self):
        """α = elementos / buckets. En chaining puede superar 1, pero mantenemos umbral para rendimiento."""
        return self.__nItems / self.celdas()

    # --- Operaciones principales ---
    def buscar(self, clave):
        """Devuelve el valor asociado o None si no existe."""
        idx = self.hash(clave)
        bucket = self.__buckets[idx]
        for k, v in bucket:
            if k == clave:
                return v
        return None

    def insertar(self, clave, valor):
        """
        Inserta o actualiza.
        - Retorna True si insertó NUEVA clave.
        - Retorna False si ACTUALIZÓ valor existente.
        Si α supera el umbral, se redimensiona y re-hashea.
        """
        idx = self.hash(clave)
        bucket = self.__buckets[idx]
        # ¿Existe la clave? -> actualizar
        for i, (k, v) in enumerate(bucket):
            if k == clave:
                bucket[i] = (clave, valor)
                return False  # actualización

        # Nueva inserción
        bucket.append((clave, valor))
        self.__nItems += 1

        if self.factorCarga() > self.__factorCargaMax:
            self.__crecimientoTabla()
        return True

    def eliminar(self, clave, ignoreMissing: bool = False):
        """Elimina la clave si existe. Lanza excepción si no existe y ignoreMissing=False."""
        idx = self.hash(clave)
        bucket = self.__buckets[idx]
        for i, (k, v) in enumerate(bucket):
            if k == clave:
                bucket.pop(i)
                self.__nItems -= 1
                return
        if not ignoreMissing:
            raise Exception(f"No se puede eliminar la clave {clave} en la Tabla Hash (no existe).")

    def recorrer(self):
        """Genera todas las (clave, valor) almacenadas (en cualquier orden)."""
        for bucket in self.__buckets:
            for par in bucket:
                yield par

    # --- Redimensionamiento ---
    def __crecimientoTabla(self):
        """Duplica tamaño (aprox.) al siguiente primo y re-inserta todo para mantener rendimiento."""
        vieja = self.__buckets
        tam = len(vieja) * 2 + 1
        while not es_primo(tam):
            tam += 2
        self.__buckets = [[] for _ in range(tam)]
        self.__nItems = 0  # se recalcula al reinsertar
        for bucket in vieja:
            for k, v in bucket:
                self.insertar(k, v)

# --- Ejemplo  de uso ---
if __name__ == "__main__":
    th = TablaHashChaining(tam=7, hash=hashSimple, factorCargaMax=0.75)
    print("celdas:", th.celdas(), "| α:", th.factorCarga())
    for k, v in [("a",1), ("b",2), ("c",3), ("d",4), ("e",5)]:
        th.insertar(k, v)
    print("buscar 'c' ->", th.buscar("c"))       # 3
    th.insertar("c", 30)                         # actualización
    print("buscar 'c' (upd) ->", th.buscar("c")) # 30
    th.eliminar("b")
    print("buscar 'b' (elim) ->", th.buscar("b"))
    print("α:", th.factorCarga(), "| items:", len(th), "| celdas:", th.celdas())
    print("contenido:", list(th.recorrer()))
```

# Uso de Tabla Hash Chaining

```python
# @title Uso TablaHashChaining

from typing import Any, List, Tuple, Generator
import ipywidgets as widgets
from IPython.display import display, Markdown, clear_output
import pandas as pd

# ------------------------------
# Hash determinista y utilidades
# ------------------------------
def hashSimple(clave: Any) -> int:
    if isinstance(clave, int):
        return clave & 0x7FFFFFFF
    if isinstance(clave, str):
        h = 0
        for ch in clave:
            h = (h * 257 + ord(ch)) & 0x7FFFFFFF
        return h
    if isinstance(clave, (tuple, list)):
        h = 0
        for x in clave:
            h = (h * 257 + hashSimple(x)) & 0x7FFFFFFF
        return h
    return hashSimple(repr(clave))

def es_primo(n: int) -> bool:
    if n <= 1: return False
    if n <= 3: return True
    if n % 2 == 0 or n % 3 == 0: return False
    i = 5
    while i * i <= n:
        if n % i == 0 or n % (i + 2) == 0:
            return False
        i += 6
    return True

# ------------------------------
# Tabla Hash con Separate Chaining
# ------------------------------
class TablaHashChaining(object):
    def __init__(self, tam: int = 7, hash=hashSimple, factorCargaMax: float = 0.75):
        if tam < 3: tam = 3
        # subir al siguiente primo para mejor dispersión
        if not es_primo(tam):
            t = tam if tam % 2 == 1 else tam + 1
            while not es_primo(t):
                t += 2
            tam = t
        self.__buckets: List[List[Tuple[Any, Any]]] = [[] for _ in range(tam)]
        self.__nItems = 0
        self.__hash = hash
        self.__factorCargaMax = factorCargaMax

    def __len__(self):
        return self.__nItems

    def celdas(self):
        return len(self.__buckets)

    def hash(self, clave):
        return self.__hash(clave) % self.celdas()

    def factorCarga(self):
        return self.__nItems / self.celdas()

    def buscar(self, clave):
        idx = self.hash(clave)
        bucket = self.__buckets[idx]
        for k, v in bucket:
            if k == clave:
                return v
        return None

    def insertar(self, clave, valor):
        idx = self.hash(clave)
        bucket = self.__buckets[idx]
        for i, (k, v) in enumerate(bucket):
            if k == clave:
                bucket[i] = (clave, valor)
                return False  # actualización
        bucket.append((clave, valor))
        self.__nItems += 1
        if self.factorCarga() > self.__factorCargaMax:
            self.__crecimientoTabla()
        return True

    def eliminar(self, clave, ignoreMissing: bool = False):
        idx = self.hash(clave)
        bucket = self.__buckets[idx]
        for i, (k, v) in enumerate(bucket):
            if k == clave:
                bucket.pop(i)
                self.__nItems -= 1
                return
        if not ignoreMissing:
            raise Exception(f"No se puede eliminar la clave {clave} (no existe).")

    def recorrer(self):
        for bucket in self.__buckets:
            for par in bucket:
                yield par

    def __crecimientoTabla(self):
        vieja = self.__buckets
        tam = len(vieja) * 2 + 1
        while not es_primo(tam):
            tam += 2
        self.__buckets = [[] for _ in range(tam)]
        self.__nItems = 0
        for bucket in vieja:
            for k, v in bucket:
                self.insertar(k, v)

# ------------------------------
# Widgets / UI
# ------------------------------
# Configuración
tam_init   = widgets.IntText(value=7, description='Tamaño:', layout=widgets.Layout(width='200px'))
alpha_max  = widgets.FloatText(value=0.75, description='α max:', layout=widgets.Layout(width='180px'))
btn_reset  = widgets.Button(description='Crear/Reset', button_style='info')

# Insertar
key_in   = widgets.Text(value='', description='Clave:',  placeholder='ej. a, 42, (1,2)')
val_in   = widgets.Text(value='', description='Valor:',  placeholder='ej. 10, "hola"')
btn_ins  = widgets.Button(description='Insertar', button_style='success')

# Buscar / Eliminar
key_find = widgets.Text(value='', description='Clave:')
btn_bus  = widgets.Button(description='Buscar')
btn_del  = widgets.Button(description='Eliminar', button_style='warning')

# Inspección de bucket
bucket_idx = widgets.BoundedIntText(value=0, min=0, max=max(0, tam_init.value-1),
                                    description='Bucket #:', layout=widgets.Layout(width='180px'))
btn_view_bucket = widgets.Button(description='Ver bucket')

# Salidas
stats_out = widgets.Output()
table_out = widgets.Output()
bucket_out = widgets.Output()
log_out   = widgets.Output()

# Estado actual
state = {'th': None}

def make_table():
    th = TablaHashChaining(tam=tam_init.value, hash=hashSimple, factorCargaMax=float(alpha_max.value))
    return th

def parse_literal(s: str):
    s = s.strip()
    if s == '':
        return ''
    import ast
    # int
    if s.lstrip('+-').isdigit():
        try: return int(s)
        except: pass
    # float
    try:
        f = float(s)
        return f
    except:
        pass
    # literal seguro (tupla/lista/dict/str con comillas)
    try:
        return ast.literal_eval(s)
    except:
        return s  # string crudo

def refresh_summary(th: TablaHashChaining, msg: str = ''):
    with stats_out:
        clear_output(wait=True)
        display(Markdown(f"**Items:** {len(th)} | **Buckets (celdas):** {th.celdas()} | **Factor de carga (α):** {th.factorCarga():.3f}"))
        if msg:
            display(Markdown(f"_Última acción:_ {msg}"))

def refresh_table(th: TablaHashChaining):
    # Construir DataFrame: por bucket, tamaño y contenido
    # (Acceso al atributo privado para docencia/visualización)
    buckets = getattr(th, "_TablaHashChaining__buckets")
    data = []
    for i, b in enumerate(buckets):
        data.append({
            "Índice": i,
            "Tamaño del bucket": len(b),
            "Contenido (pares)": list(b)  # lista de (clave, valor)
        })
    df = pd.DataFrame(data)
    with table_out:
        clear_output(wait=True)
        display(df.style.hide(axis='index'))

    # ajustar rango del selector de bucket
    bucket_idx.max = len(buckets) - 1
    if bucket_idx.value > bucket_idx.max:
        bucket_idx.value = bucket_idx.max

def view_single_bucket(th: TablaHashChaining, idx: int):
    buckets = getattr(th, "_TablaHashChaining__buckets")
    if idx < 0 or idx >= len(buckets):
        with bucket_out:
            clear_output(wait=True)
            print("Índice de bucket fuera de rango.")
        return
    contenido = buckets[idx]
    df = pd.DataFrame(contenido, columns=["Clave", "Valor"]) if contenido else pd.DataFrame(columns=["Clave","Valor"])
    with bucket_out:
        clear_output(wait=True)
        display(Markdown(f"**Bucket #{idx}** — tamaño: {len(contenido)}"))
        display(df if not df.empty else Markdown("_(vacío)_"))

# Handlers
def on_reset(_):
    th = make_table()
    state['th'] = th
    with log_out:
        clear_output(wait=True)
        print(f"Tabla creada: tamaño={th.celdas()} | α max={alpha_max.value}")
    refresh_summary(th, "Tabla creada/reset")
    refresh_table(th)
    view_single_bucket(th, bucket_idx.value)

def on_insert(_):
    th = state.get('th') or make_table()
    state['th'] = th
    clave = parse_literal(key_in.value)
    valor = parse_literal(val_in.value)
    try:
        nuevo = th.insertar(clave, valor)
        msg = f"Insertar clave={repr(clave)}, valor={repr(valor)} " + ("(nuevo)" if nuevo else "(actualización)")
        with log_out:
            clear_output(wait=True)
            print(msg)
        refresh_summary(th, msg)
        refresh_table(th)
        view_single_bucket(th, th.hash(clave))
    except Exception as e:
        with log_out:
            clear_output(wait=True)
            print("ERROR:", e)

def on_buscar(_):
    th = state.get('th')
    if th is None:
        with log_out:
            clear_output(wait=True)
            print("Primero crea la tabla.")
        return
    clave = parse_literal(key_find.value)
    val = th.buscar(clave)
    with log_out:
        clear_output(wait=True)
        print(f"Buscar clave={repr(clave)} ->", val)
    if val is not None:
        view_single_bucket(th, th.hash(clave))

def on_eliminar(_):
    th = state.get('th')
    if th is None:
        with log_out:
            clear_output(wait=True)
            print("Primero crea la tabla.")
        return
    clave = parse_literal(key_find.value)
    try:
        th.eliminar(clave)
        with log_out:
            clear_output(wait=True)
            print(f"Eliminar clave={repr(clave)} -> OK")
        refresh_summary(th, f"Eliminar clave={repr(clave)}")
        refresh_table(th)
        view_single_bucket(th, th.hash(clave))
    except Exception as e:
        with log_out:
            clear_output(wait=True)
            print("ERROR:", e)

def on_view_bucket(_):
    th = state.get('th')
    if th is None:
        with log_out:
            clear_output(wait=True)
            print("Primero crea la tabla.")
        return
    view_single_bucket(th, int(bucket_idx.value))

# Bind
btn_reset.on_click(on_reset)
btn_ins.on_click(on_insert)
btn_bus.on_click(on_buscar)
btn_del.on_click(on_eliminar)
btn_view_bucket.on_click(on_view_bucket)

# Layout
display(Markdown("## Tabla Hash (Separate Chaining) — Panel Interactivo"))
display(Markdown("### Configuración"))
display(widgets.HBox([tam_init, alpha_max, btn_reset]))

display(Markdown("### Insertar"))
display(widgets.HBox([key_in, val_in, btn_ins]))

display(Markdown("### Buscar / Eliminar"))
display(widgets.HBox([key_find, btn_bus, btn_del]))

display(Markdown("### Resumen"))
display(stats_out)

display(Markdown("### Buckets (tamaño y contenido)"))
display(table_out)

display(Markdown("### Inspección puntual de bucket"))
display(widgets.HBox([bucket_idx, btn_view_bucket]))
display(bucket_out)

display(Markdown("### Log"))
display(log_out)

# inicializa
on_reset(None)
```

## Conclusiones: Open Addressing vs Separate Chaining

- **Uso de memoria**  
  - *Open addressing*: eficiente, solo usa el arreglo principal.  
  - *Separate chaining*: necesita memoria extra (listas/nodos).  

- **Factor de carga**  
  - *Open addressing*: funciona bien con α bajo (≈ 0.5–0.7).  
  - *Separate chaining*: tolera α > 1, porque los buckets pueden crecer.  

- **Rendimiento promedio**  
  - *Open addressing*: muy rápido si la tabla está poco cargada.  
  - *Separate chaining*: rendimiento más estable con alta carga, aunque se recorra el bucket.  

- **Complejidad de eliminación**  
  - *Open addressing*: requiere “marcadores de tumba”.  
  - *Separate chaining*: basta eliminar de la lista del bucket.  

- **Acceso a memoria**  
  - *Open addressing*: más *cache-friendly* (datos contiguos).  
  - *Separate chaining*: menos eficiente en memoria, pero más flexible ante colisiones.
